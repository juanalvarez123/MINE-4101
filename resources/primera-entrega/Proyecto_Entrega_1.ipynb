# ****Validación de Hipótesis****

###1. Las mujeres son las mayores víctimas en robos

```python
genero_df = med_hurto_df.groupby(['sexo'])['sexo'].count()
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99','#c2c2f0','#ffb3e6','#ADFF2F','#00CCFF','#FF9966','#FF7F50','#00FF00','#FF1493','#696969','#A0522D']
ax = genero_df.plot(kind='pie', rot=0, autopct="%.2f",figsize=(12, 6), colors=colors)
font = {'family':'serif','color':'darkred','size':25}
plt.title("Porcentaje de robos por género", font)
ax.legend(bbox_to_anchor=(1.01, 1.02), loc='upper left', fontsize=15)
plt.style.use('seaborn')
plt.tick_params(labelsize=14)
plt.show()
```

De acuedo a los datos de la alcaldía de Medellín y al anterior gráfico las mujeres no son las mayores víctimas de los robos sino son los hombres con un 58.30%, por lo tanto se rechaza la hipótesis.

###2. La mayoría de hurtos ocurren a personas que iban caminando.


```python
transporte_df = med_hurto_df.groupby(['medio_transporte'])['medio_transporte'].count()
ax = transporte_df.plot(kind='pie', rot=0, autopct="%.2f",figsize=(18, 9), colors=colors)
font = {'family':'serif','color':'darkred','size':25}
plt.title("Porcentaje de robos por modalidad de transporte", font)
ax.legend(bbox_to_anchor=(1.01, 1.02), loc='upper left', fontsize=15)
plt.style.use('seaborn')
plt.tick_params(labelsize=14)
plt.show()
```

De acuerdo al gráfico se puede ver que el 70.64% de los hurtos ocurren cuando las personas se encuentran caminando y los siguientes dos caso con mayor porcentaje es en el vehículo con 10.14% y el 3.92% en motocicleta con parrillero, el cual tambien debe presentarse en las calles. ### 3. Las zonas donde ocurren más hurtos están ubicados en el centro de la ciudad.

Por medio del siguiente diccionario se distribuyeron las diferentes comunas en sus respectivas zonas.

```python
zonas ={
    'Nororiente': ['1', '2', '3', '4'],
    'Noroccidental': ['5', '6', '7'],
    'Centro Oriental': ['8', '9', '10'],
    'Centro Occidental': ['11', '12', '13'],
    'Suroriental': ['14'],
    'Suroccidental': ['15', '16'],
    'Corregimientos': ['50', '60', '70', '80', '90'],
} 
```

```python
hurtos_nororiente = med_hurto_df[med_hurto_df['codigo_comuna'].isin(zonas['Nororiente'])]
hurtos_noroccidente = med_hurto_df[med_hurto_df['codigo_comuna'].isin(zonas['Noroccidental'])]
hurtos_centro_oriente = med_hurto_df[med_hurto_df['codigo_comuna'].isin(zonas['Centro Oriental'])]
hurtos_centro_occidente = med_hurto_df[med_hurto_df['codigo_comuna'].isin(zonas['Centro Occidental'])]
hurtos_suroriente = med_hurto_df[med_hurto_df['codigo_comuna'].isin(zonas['Suroriental'])]
hurtos_suroccidente = med_hurto_df[med_hurto_df['codigo_comuna'].isin(zonas['Suroccidental'])]
hurtos_corregimientos = med_hurto_df[med_hurto_df['codigo_comuna'].isin(zonas['Corregimientos'])]

data = [
    hurtos_nororiente.shape[0], 
    hurtos_noroccidente.shape[0], 
    hurtos_centro_oriente.shape[0], 
    hurtos_centro_occidente.shape[0], 
    hurtos_suroriente.shape[0], 
    hurtos_suroccidente.shape[0], 
    hurtos_corregimientos.shape[0]
  ]
keys = zonas.keys()
explode = [0, 0, 0.1, 0.1, 0, 0, 0]
font = {'family':'serif','color':'darkred','size':25}
plt.title("Porcentaje de hurtos por zona", font)
plt.pie(data, labels=keys,
        explode=explode, autopct='%.0f%%', colors=colors)
plt.show()
```

En este gráfico se confirma la hipótesis de que los barrios con más hurtos se encuentran en el centro de la ciudad, puesto que sumando los 2 sectores centrales se encuentran 55% de los hurtos que ocurren en Medellín y sus alrededores.

### 5. Los viernes y los sábados en horas de la noche son los momentos más peligrosos y donde se presenta el mayor hurto de celulares.

```python
med_hurto_df['dayofweek'] = pd.DatetimeIndex(med_hurto_df['fecha_hecho']).dayofweek
med_hurto_df['hour'] = pd.DatetimeIndex(med_hurto_df['fecha_hecho']).hour

med_hurto_df['dayofweek'] = med_hurto_df['dayofweek'].astype(str)
med_hurto_df['dayofweek'] = med_hurto_df['dayofweek'].str.replace('0','lunes')
med_hurto_df['dayofweek'] = med_hurto_df['dayofweek'].str.replace('1','martes')
med_hurto_df['dayofweek'] = med_hurto_df['dayofweek'].str.replace('2','miercoles')
med_hurto_df['dayofweek'] = med_hurto_df['dayofweek'].str.replace('3','jueves')
med_hurto_df['dayofweek'] = med_hurto_df['dayofweek'].str.replace('4','viernes')
med_hurto_df['dayofweek'] = med_hurto_df['dayofweek'].str.replace('5','sabado')
med_hurto_df['dayofweek'] = med_hurto_df['dayofweek'].str.replace('6','domingo')
hurtos_por_dia = med_hurto_df.groupby(['dayofweek']).size().sort_values()
ax = hurtos_por_dia.plot(kind='pie', rot=0, autopct="%.2f",figsize=(18, 9), colors=colors)
font = {'family':'serif','color':'darkred','size':25}
plt.title("Porcentaje de hurtos por día de la semana", font)
ax.legend(bbox_to_anchor=(1.01, 1.02), loc='upper left', fontsize=15)
plt.tick_params(labelsize=14)
plt.show()
```

En este gráfico se puede observar que efectivamente los dias viernes y sábado son los dias que más robos ocurren, aunque con poco margen respecto a los miércoles y jueves. Aunque todavia falta observar la ocurrencia de esto si es de día o de noche.

```python
med_hurto_df['isnight'] = np.where((med_hurto_df.hour >= 18) | (med_hurto_df.hour <= 5), 'noche', 'día')

hurtos_por_dia = med_hurto_df.groupby(['dayofweek', 'isnight']).size().sort_values()
ax = hurtos_por_dia.plot(kind='pie', rot=0, autopct="%.2f",figsize=(18, 9), colors=colors)
font = {'family':'serif','color':'darkred','size':25}
plt.title("Porcentaje de hurtos por día de la semana", font)
ax.legend(bbox_to_anchor=(1.01, 1.02), loc='upper left', fontsize=15)
plt.tick_params(labelsize=14)
plt.show()
```

Aqui podemos observar una tendencia sorprendente, ocurren más robos desde las 5am hasta las 6pm que en la noche. Otra cosa a tomar en cuenta es que no necesariamente los dias viernes o sábados ocurren más robos, ya que los miércoles de día se presentaron más robos.

Como una posible explicación a los robos durante el día, se realizó un analisis detallado a continuación:

```python
med_hurto_df['isnight'] = np.where((med_hurto_df.hour >= 12), '12:00 - 23:59', '00:00 - 11:59')

hurtos_por_dia = med_hurto_df.groupby(['dayofweek', 'isnight']).size().sort_values()
ax = hurtos_por_dia.plot(kind='pie', rot=0, autopct="%.2f",figsize=(18, 9), colors=colors)
font = {'family':'serif','color':'darkred','size':25}
plt.title("Porcentaje de hurtos por día de la semana", font)
ax.legend(bbox_to_anchor=(1.01, 1.02), loc='upper left', fontsize=15)
plt.tick_params(labelsize=14)
plt.show()
```

Se puede afirmar que la mayoría de los robos estan ocurriendo en la tarde y la noche, y en este caso se puede observar una diferencia más notoria entre los hurtos del viernes y el resto de dias. Por lo tanto se puede concluir que el viernes entre las horas de la tarde y la noche es el momento donde más hurtos ocurren, por otro lado el sábado no es tan peligroso como se afirmaba en la hipótesis

###1. ¿Cúal es la relación de los bienes robados entre adolescentes, jovenes, adultos y adultos mayores?

```python
med_hurto_edad_df = med_hurto_df[(med_hurto_df['edad']>12)]
med_hurto_edad_df['grupo-edad'] = pd.cut(med_hurto_df['edad'], [13, 20, 40, 60, 100], labels=['13-20', '21-40', '41-60', '61-100'])
med_hurto_edad_df = med_hurto_edad_df.groupby(['grupo-edad','bien']).agg({'bien':['count']})
med_hurto_edad_df = med_hurto_edad_df.sort_values(['grupo-edad',('bien', 'count')], ascending=[True, False]).groupby('grupo-edad').head(5)
med_hurto_edad_df.reset_index(inplace=True)
med_hurto_edad_df.columns = ['grupo-edad','bien','count']
```

```python
dfp = med_hurto_edad_df.pivot(index='grupo-edad', columns='bien', values='count')

ax = dfp.plot(kind='bar', rot=0, figsize=(24, 12))
font = {'family':'serif','color':'darkred','size':15}
plt.xlabel("Years", font)
plt.ylabel("Top", font)
plt.title("Variación de robos por grupos de edad", font)
ax.legend(bbox_to_anchor=(1.01, 1.02), loc='upper left', fontsize=14)
plt.style.use('seaborn')
plt.tick_params(labelsize=14)
plt.show()
```

De acuerdo a los diferentes grupos de población en cada una de estas el bien más robado es el celular, seguido del efectivo.

### 2. ¿Cuáles son los barrios/zonas más peligrosas de Medellín en las cuales se requiere mayor presencia policial?

### 3. ¿Cómo ha cambiado la cantidad de hurtos de las diferentes comunas en el tiempo?

```python
med_hurto_df['year'] = pd.DatetimeIndex(med_hurto_df['fecha_hecho']).year
hurtos_por_year_comuna = med_hurto_df.groupby(['year', 'codigo_comuna']).size()
min_year = med_hurto_df.year.min()
max_year = med_hurto_df.year.max()

data = {}

num = max_year - min_year + 1

for sector in zonas.keys():
  data[sector] = [0] * num


for ind in hurtos_por_year_comuna.index:
  for sector in zonas.keys():
    if ind[1] in zonas[sector]:
      data[sector]
      data[sector][ind[0]-min_year] += hurtos_por_year_comuna[ind]


colors = plt.rcParams['axes.prop_cycle'].by_key()['color']
fig = plt.figure(figsize=(15, 10))


for i, (name, values) in enumerate(data.items()):
  sns.lineplot(x=[str(year) for year in range(min_year, max_year+1)], y=values, alpha=0.9, color=colors[i % len(colors)], marker='o')

x_pos = np.arange(max_year-min_year+1)
plt.xticks(x_pos, [str(year) for year in range(min_year, max_year+1)])

plt.legend(data.keys(), loc=(1.04, 0))

plt.show()
```

En este gráfico se puede observar que la tendencia para todas las comunas es muy similar, estuvo bajo desde el año 2003 al 2012 y entre los años 2013 a 2019 hubo un aumento exagerado de los robos. Existen 2 posibles expicaciones a este fenómeno:


*   Que el proceso de almacenamiento de casos de hurtos no estuviera bien implementado antes de 2013.
*   Aproximadamente en 2014 fue la segunda ola de migraciones masivas de Venezolanos a Colombia.

Es difícil concluir cual de estas 2 opciones es correcta con los datos que tenemos actualmente.

Por último, podemos observar que desde el año 2019 los robos se disminuyeron considerablemente.
